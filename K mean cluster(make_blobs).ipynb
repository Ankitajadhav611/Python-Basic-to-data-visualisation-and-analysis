K Mean cluster 

Working on an aritificial data developed using make_blobs function of scikit learn

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

Useful link to generate the test data set:https://machinelearningmastery.com/generate-test-datasets-python-scikit-learn/
https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html

```python
from sklearn.datasets import make_blobs
```

```python
#create a data
data  = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
data[0]
```

this is a tuple with sample , where sample has two feature and thats default

```python
sns.scatterplot(x=data[0][:,0],y=data[0][:,1],hue=data[1],palette='rainbow')
```

creating a cluster

```python
from sklearn.cluster import KMeans
```

```python
Kmeans = KMeans(n_clusters=4) #since we have already declared k= 4 while creating the dataset 
```

```python
Kmeans.fit(data[0])
```

```python
Kmeans.cluster_centers_
```

```python
Kmeans.labels_
```

```python
sns.set_style('whitegrid')
fig,(ax1,ax2) = plt.subplots(1,2,sharey=True,figsize=(10,6))
ax1.set_title('K means')
ax1.scatter(data[0][:,0],data[0][:,1],c=Kmeans.labels_,cmap='rainbow')
ax2.set_title('Original data set')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
